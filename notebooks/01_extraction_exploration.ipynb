# Extraction Exploration
Walk through parsing a synthetic sample form and inspecting the inferred schema + extracted fields.

Requires `OPENAI_API_KEY` set.

```python
from pathlib import Path
from form_agent.agent import FormAgent

agent = FormAgent()
sample = Path('../data/samples/job_app_01.pdf')
form = agent.ingest(sample)
form.id, form.form_type, list(form.fields)
```

```python
import json
print(json.dumps(form.fields, indent=2, default=str))
```

```python
ans = agent.ask(form.id, 'What is the candidate\'s most recent employer?')
print(ans.answer, '->', ans.confidence)
for c in ans.citations: print(' ', c)
```

```python
summary = agent.summarize(form.id)
summary
```
